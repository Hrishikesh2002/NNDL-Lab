```python
## Write your code here

from tensorflow import keras
from keras import layers

import numpy as np
import matplotlib.pyplot as plt

# This is the size of our encoded representations
encoding_dim = 32  # 32 floats -> compression of factor 24.5, assuming the input is 784 floats

# This is our input image
input_img = keras.Input(shape=(784,))
# "encoded" is the encoded representation of the input
encoded = layers.Dense(encoding_dim, activation='relu')(input_img)
# "decoded" is the lossy reconstruction of the input
decoded = layers.Dense(784, activation='sigmoid')(encoded)

# This model maps an input to its reconstruction
autoencoder = keras.Model(input_img, decoded)
```

```python
# This model maps an input to its encoded representation
encoder = keras.Model(input_img, encoded)
```

```python
# This is our encoded (32-dimensional) input
encoded_input = keras.Input(shape=(encoding_dim,))
# Retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]
# Create the decoder model
decoder = keras.Model(encoded_input, decoder_layer(encoded_input))
```

```python
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
print(autoencoder.summary())
```

```python
from keras.datasets import mnist
import numpy as np
(x_train, _), (x_test, _) = mnist.load_data()
x_val = x_test[len(x_test)//2:]
x_test = x_test[:len(x_test)//2]

# Print the shape of the training data
print(x_train.shape)


# Print the shape of the test data
print(x_test.shape)

# Take 6000 random samples from the training data 
# First generate random indices
random_indices = np.random.choice(range(len(x_train)), 6000, replace=False)
# Use the random indices to select the samples
x_train = x_train[random_indices]


# Take 1000 random samples from the first half of the test data
# First generate random indices
random_indices = np.random.choice(range(len(x_test)), 1000, replace=False)
# Use the random indices to select the samples
x_test = x_test[random_indices]


# Generate validation data from the second half of the test data; take 1000 samples randomly
random_indices = np.random.choice(range(len(x_val)), 1000, replace=False)
x_val = x_val[random_indices]


# Display one of the images
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
plt.show()

# Use unaltered images for training and noisy images for testing (add noise to the images)

# Add noise to the images
noise_factor = 50

x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)

# Clip the images to be between 0 and 255
x_test_noisy = np.clip(x_test_noisy, 0., 255.)


x_val_noisy = x_val + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_val.shape)

# Clip the images to be between 0 and 255
x_val_noisy = np.clip(x_val_noisy, 0., 255.)

# Display the noisy images along with the original images

n = 10  # How many digits we will display

plt.figure(figsize=(20, 4))

for i in range(n):
    # Display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i])
    plt.title('Original')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display noisy
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(x_test_noisy[i])
    plt.title('Noisy')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
plt.show()

# Show just one image and its noisy version
plt.imshow(x_test[0])
plt.show()
plt.imshow(x_test_noisy[0])
plt.show()





x_train = x_train.astype('float32') / 255.
x_test_noisy = x_test_noisy.astype('float32') / 255.
x_val_noisy = x_val_noisy.astype('float32') / 255.

x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test_noisy = x_test_noisy.reshape((len(x_test), np.prod(x_test_noisy.shape[1:])))
x_val_noisy = x_val_noisy.reshape(len(x_val_noisy), np.prod(x_val_noisy.shape[1:]))

x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
x_val = x_val.reshape((len(x_val), np.prod(x_val.shape[1:])))

print(x_train.shape)
print(x_test_noisy.shape)
print(x_val_noisy.shape)

print(x_test.shape)
print(x_val.shape)
```

```python
# Train the autoencoder on the training data

print(x_train.shape)
print(x_val.shape)
print(len(x_train[1]))
autoencoder.fit(x_train, x_train,
                batch_size=70,
                epochs=100,
                shuffle=True,
                validation_data=(x_val, x_val))
```

```python
# Display one of the noisy images
plt.imshow(x_test_noisy[0].reshape(28, 28))
```

```python
# Encode and decode some digits
# Test the autoencoder on the noisy test data
encoded_imgs = encoder.predict(x_test_noisy)
decoded_imgs = decoder.predict(encoded_imgs)

# Display the noisy images, encoded images, and the reconstructed images
n = 10  # How many digits we will display
plt.figure(figsize=(20, 4))
# Don't reshape the images back to 28x28, as the input to the autoencoder is 784-dimensional
for i in range(n):
    # Display noisy
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.title('Noisy')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display encoded
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(encoded_imgs[i].reshape(8, 4))
    plt.title('Encoded')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display reconstructed
    ax = plt.subplot(3, n, i + 1 + 2*n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.title('Reconstructed')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

```

```python
# Implement a deep autoencoder

# Create a denoising autoencoder
input_img = keras.Input(shape=(784,))
encoded_2 = layers.Dense(128, activation='relu')(input_img)
encoded_2 = layers.Dense(64, activation='relu')(encoded_2)
encoded_2 = layers.Dense(32, activation='relu')(encoded_2)

decoded_2 = layers.Dense(64, activation='relu')(encoded_2)
decoded_2 = layers.Dense(128, activation='relu')(decoded_2)
decoded_2 = layers.Dense(784, activation='sigmoid')(decoded_2)

autoencoder_2 = keras.Model(input_img, decoded_2)
autoencoder_2.compile(optimizer='adam', loss='binary_crossentropy')


print(autoencoder_2.summary())
```

```python
# Get the encoder and decoder models
encoder_2 = keras.Model(input_img, encoded_2)
encoded_input = keras.Input(shape=(encoding_dim,))

decoder_2 = keras.Model(encoded_input, autoencoder_2.layers[-1](autoencoder_2.layers[-2](autoencoder_2.layers[-3](encoded_input))))
```

```python
# Train the deep autoencoder on the training data
autoencoder_2.fit(x_train, x_train,
                epochs=100,
                shuffle=True,
                validation_data=(x_val, x_val))

```

```python
# Test the deep autoencoder on the noisy test data
encoded_imgs_shallow = encoder.predict(x_test_noisy)
encoded_imgs_deep = encoder_2.predict(x_test_noisy)
decoded_imgs_shallow = decoder.predict(encoded_imgs_shallow)
decoded_imgs_deep = decoder_2.predict(encoded_imgs_deep)

# Display the noisy images, encoded images, and the reconstructed images
n = 10  # How many digits we will display

plt.figure(figsize=(20, 4))

for i in range(n):
    # Display noisy
    ax = plt.subplot(4, n, i + 1)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.title('Noisy')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display shallow reconstructed
    ax = plt.subplot(4, n, i + 1 + n)
    plt.imshow(decoded_imgs_shallow[i].reshape(28, 28))
    plt.title('Shallow Rec.')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display deep reconstructed
    ax = plt.subplot(4, n, i + 1 + 2*n)
    plt.imshow(decoded_imgs_deep[i].reshape(28, 28))
    plt.title('Deep Rec.')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display original
    ax = plt.subplot(4, n, i + 1 + 3*n)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.title('Original')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
plt.show()

```

```python
# Use standard metrics to compare the performance of the shallow and deep autoencoders for image reconstruction

from sklearn.metrics import mean_squared_error
from skimage.metrics import structural_similarity as ssim

# Calculate the mean squared error between the original and reconstructed images
mse_shallow = mean_squared_error(x_test, decoded_imgs_shallow)
mse_deep = mean_squared_error(x_test, decoded_imgs_deep)

print("Mean Squared Error for Shallow Autoencoder: ", mse_shallow)
print("Mean Squared Error for Deep Autoencoder: ", mse_deep)

# Use other metrics to compare the performance of the shallow and deep autoencoders for image reconstruction

# Calculate the SSIM between the original and reconstructed images
# Loop through the images and calculate the SSIM

ssim_shallow = 0
ssim_deep = 0

for i in range(len(x_test)):
    ssim_shallow += ssim(x_test_noisy[i], decoded_imgs_shallow[i], data_range=1)
    ssim_deep += ssim(x_test_noisy[i], decoded_imgs_deep[i], data_range=1)
    
ssim_shallow /= len(x_test)
ssim_deep /= len(x_test)

print("\nSSIM for Shallow Autoencoder: ", ssim_shallow)
print("SSIM for Deep Autoencoder: ", ssim_deep)

```

## Inferences and Conclusion : State all the key observations and conclusion

1. Autoencoders generate a compressed representation of the input data.

2. The compressed representation is then used to reconstruct the input data.

3. More layers help in better representation of the input data, so the reconstructed image in more accurate (more SSIM value).

```python
## Write your code here

# Implement regularized autoencoder

# Create a regularized autoencoder

from keras import regularizers

input_img = keras.Input(shape=(784,))
encoded_reg = layers.Dense(128, activation='relu', activity_regularizer=regularizers.l1(10e-5))(input_img)
encoded_reg = layers.Dense(64, activation='relu', activity_regularizer=regularizers.l1(10e-5))(encoded_reg)
encoded_reg = layers.Dense(32, activation='relu', activity_regularizer=regularizers.l1(10e-5))(encoded_reg)

decoded_reg = layers.Dense(64, activation='relu', activity_regularizer=regularizers.l1(10e-5))(encoded_reg)
decoded_reg = layers.Dense(128, activation='relu', activity_regularizer=regularizers.l1(10e-5))(decoded_reg)
decoded_reg = layers.Dense(784, activation='sigmoid', activity_regularizer=regularizers.l1(10e-5))(decoded_reg)

autoencoder_reg = keras.Model(input_img, decoded_reg)
autoencoder_reg.compile(optimizer='adam', loss='binary_crossentropy')

print(autoencoder_reg.summary())
```

```python
# Get the encoder and decoder models
encoder_reg = keras.Model(input_img, encoded_reg)
encoded_input = keras.Input(shape=(encoding_dim,))

decoder_reg = keras.Model(encoded_input, autoencoder_reg.layers[-1](autoencoder_reg.layers[-2](autoencoder_reg.layers[-3](encoded_input))))

```

```python
# Train the regularized autoencoder on the training data
autoencoder_reg.fit(x_train, x_train,
                epochs=100,
                shuffle=True,
                validation_data=(x_val, x_val))

```

```python
# Test the regularized autoencoder on the noisy test data

encoded_imgs_reg = encoder_reg.predict(x_test_noisy)
decoded_imgs_reg = decoder_reg.predict(encoded_imgs_reg)



# Display the noisy images, encoded images, and the reconstructed images for regularised and deep networds
n = 10  # How many digits we will display

plt.figure(figsize=(20, 4))

for i in range(n):
    # Display noisy
    ax = plt.subplot(4, n, i + 1)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.title('Noisy')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display regularized reconstructed
    ax = plt.subplot(4, n, i + 1 + n)
    plt.imshow(decoded_imgs_reg[i].reshape(28, 28))
    plt.title('Regularized Rec.')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display deep reconstructed
    ax = plt.subplot(4, n, i + 1 + 2*n)
    plt.imshow(decoded_imgs_deep[i].reshape(28, 28))
    plt.title('Deep Rec.')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display original
    ax = plt.subplot(4, n, i + 1 + 3*n)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.title('Original')
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
plt.show()
```

```python
# Compare the performance of the regularized autoencoder with the deep autoencoder 

# Calculate the mean squared error between the original and reconstructed images
mse_reg = mean_squared_error(x_test, decoded_imgs_reg)
mse_deep = mean_squared_error(x_test, decoded_imgs_deep)

print("Mean Squared Error for Regularized Autoencoder: ", mse_reg)
print("Mean Squared Error for Deep Autoencoder: ", mse_deep)

# Calculate the SSIM between the original and reconstructed images
# Loop through the images and calculate the SSIM

ssim_reg = 0
ssim_deep = 0

for i in range(len(x_test)):
    ssim_reg += ssim(x_test_noisy[i], decoded_imgs_reg[i], data_range=1)
    ssim_deep += ssim(x_test_noisy[i], decoded_imgs_deep[i], data_range=1)
    
ssim_reg /= len(x_test)
ssim_deep /= len(x_test)
ssim_shallow /= len(x_test)

print("\nSSIM for Regularized Autoencoder: ", ssim_reg)
print("SSIM for Deep Autoencoder: ", ssim_deep)
```
